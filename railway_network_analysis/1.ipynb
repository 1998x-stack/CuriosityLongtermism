```python
import pandas as pd
df = pd.read_excel('/Users/mx/Desktop/CuriosityLongtermism/railway_network_analysis/全国火车站点.xlsx')
df.head()
```

```python
import geopandas as gpd
import matplotlib.pyplot as plt
from shapely.geometry import Point
```

```python
# 3. Create GeoDataFrame from your DataFrame
geometry = [Point(xy) for xy in zip(df.经度, df.纬度)]
geo_df = gpd.GeoDataFrame(df, geometry=geometry, crs="EPSG:4326")
```

```python
# 4. Create plot
fig, ax = plt.subplots(figsize=(15, 15))

# Plot railway stations
geo_df.plot(ax=ax,
            markersize=8,
            color='red',
            marker='o',
            label='Railway Stations')


# Customize plot
plt.title('China Railway Stations Distribution', fontsize=20)
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)

# Set appropriate bounds (mainland China)
plt.xlim(73, 135)
plt.ylim(15, 55)

plt.show()
```
